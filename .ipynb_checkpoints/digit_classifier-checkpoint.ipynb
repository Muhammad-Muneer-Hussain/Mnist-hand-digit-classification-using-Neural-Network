```python
from tensorflow import keras
from tensorflow.keras import saving
```

```python
model= saving.load_model('hand_written_digit_classifier.keras')
```

```python
import numpy as np
from PIL import Image
```

```python
def pred(image_path):
    image = Image.open(image_path).convert('L')
    image = image.resize((28,28))
    img_arr = np.array(image).astype('float32')/255
    img_arr = img_arr.reshape((1,28*28))
    prediction = model.predict(img_arr)
    predicted_label = np.argmax(prediction)
    print(f'Predicted digit is : {predicted_label}')
```

```python
pred('D:\\mnist\\Mnist-hand-digit-classification-using-Neural-Network\\three.png')
```
